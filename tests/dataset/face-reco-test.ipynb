```python
def verify(img1_path, img2_path):
    img1 = cv2.imread(img1_path)
    img2 = cv2.imread(img2_path)
    plt.imshow(img1[:, :, ::-1])
    plt.show()
    plt.imshow(img2[:, :, ::-1])
    plt.show()
    result = DeepFace.verify(img1_path,img2_path)
    #print("result: ", result)
    
    verification = result['verified']
    if(verification):
        print("It is Same Image")
    else:
        print("It is different Image")
```

```python
verify("abhinav0.jpeg","img1.jpg")
```
